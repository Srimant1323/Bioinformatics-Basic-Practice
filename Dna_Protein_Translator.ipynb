```python
from Bio.Seq import Seq
import matplotlib.pyplot as plt
```

```python
def analyze_dna(sequence):

    sequence = sequence.upper()

    print("Original DNA Sequence:")
    print(sequence)
    print("\nLength:", len(sequence))

    # GC Content
    g = sequence.count("G")
    c = sequence.count("C")
    gc_content = ((g + c) / len(sequence)) * 100
    print("GC Content: {:.2f}%".format(gc_content))

    # Transcription (DNA -> mRNA)
    mrna = sequence.replace("T", "U")
    print("\nmRNA Sequence:")
    print(mrna)

    # Translation (DNA -> Protein)
    dna_seq = Seq(sequence)
    protein = dna_seq.translate(to_stop=True)
    print("\nProtein Sequence:")
    print(protein)

    return gc_content, protein
```
